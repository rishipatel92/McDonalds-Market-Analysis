```python
df_2 = df.drop(["Like","Age","VisitFrequency","Gender"],axis=1)
```

```python
df_2
```

```python
from sklearn.decomposition import PCA
from sklearn import preprocessing
```

```python
pca_data = preprocessing.scale(df_2)
pca = PCA(n_components=8)
pc = pca.fit_transform(df_2)
names = ['pc1','pc2','pc3','pc4','pc5','pc6','pc7','pc8']
dx = pd.DataFrame(data = pc, columns = names)
dx
```

```python
loadings = pca.components_
num_pc = pca.n_features_
pc_list = ["PC"+str(i) for i in list(range(1, num_pc+1))]
loadings_df = pd.DataFrame.from_dict(dict(zip(pc_list, loadings)))
loadings_df['variable'] = df_2.columns.values
loadings_df = loadings_df.set_index('variable')
loadings_df
```

```python
pt.figure(figsize=(15,6),dpi=100)
ax = sns.heatmap(loadings_df, annot=True, cmap='Spectral')
pt.show()
```

```python
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
model = KMeans()
visualizer = KElbowVisualizer(model, k=(1,8)).fit(dx)
visualizer.show()
```

```python
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=0).fit(dx)
df['cluster_num'] = kmeans.labels_
```

```python
sns.scatterplot(data=dx, x="pc1", y="pc2", hue=kmeans.labels_)
pt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], 
            marker="X", c="r", s=80, label="centroids")
pt.legend()
pt.show()
```
